# ANALISANDO A TENSÃO COM UM DIODO ZENER

Importando as bibliotecas necessárias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Lendo a planilha e armazenando as colunas em varíáveis

```python
df = pd.read_csv('diodo.csv')
print(df)
tensao_fonte = df['Tensao da fonte']
tensao_saida = df['Tensao da fonte (V ou mV)']
```

Processando os dados que vieram com letras na tensão do resistor

```python
tensao_resistor = list()
for valor in list(df['Tensao no resistor (V ou mV)']):
    if "mV" in valor:
        valor = valor.replace("mV","")
        valor = float(float(valor) / 1000)
    elif "V" in valor:
        valor = float(valor.replace("V",""))
    tensao_resistor.append(valor)
print(tensao_resistor)
```

Plotando o gráfico da tensão de saída

```python
sns.set_style('darkgrid')
plt.xlabel('Tensão da Fonte')
plt.ylabel('Tensão de Saída')
sns.lineplot(x=tensao_fonte, y=tensao_saida, color = 'blue', marker='o')
plt.savefig('tensao_saida.png')
plt.show()
```

Plotando o grafico de tensão no resistor

```python
sns.set_style('darkgrid')
plt.xlabel('Tensão da Fonte')
plt.ylabel('Tensão no Resistor')
plt.plot(tensao_fonte,tensao_resistor, marker='o', color = 'red')
plt.savefig('tensao_resitor.png')
plt.show()
```
